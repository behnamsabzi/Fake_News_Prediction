# 0 Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder

import nltk
import re
import string
import os

# Download NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

# 1 Load and Label the Datasets

First we define the file paths for the fake and real news datasets stored locally.

- The datasets are read into two separate pandas DataFrames:
    - fake_df contains fake news articles
    - true_df contains real news articles

## 1.1 Add Labels
- A new column named label is added to each DataFrame:
    - 0 is used to label fake news
    - 1 is used to label real news

**This creates the target variable for our classification task.**

## 1.2 Standardize subject labels

To ensure a fair and consistent evaluation of model performance across different content domains, we need to standardize the subject labels between the fake and real news datasets.

The fake news dataset contains a more granular set of subjects (e.g., 'left-news', 'US_News', 'Middle-east') that do not directly match the broader categories used in the true news dataset (e.g., 'politicsNews', 'worldnews'). To address this mismatch, we map related subjects in the fake news data to match the domain labels used in the true news data.

We group the subjects as follows:

- **'politics'**, **'Government News'**, and **'left-news'** are mapped to **politicsNews**

- **'News'**, **'US_News',** and **'Middle-east'** are mapped to **worldnews**

This standardization step allows us to reliably analyze the model's classification performance within each shared domain.


## 1.3 Combine Datasets

Combines the two datasets into a single DataFrame called **df** for easier processing and modeling.

- **ignore_index=True** ensures that the index is reset after merging.

## 1.4 Quick Look at Data

```python
# Path to files
fake_path = "Fake.csv"
true_path = "True.csv"

fake_df = pd.read_csv(fake_path)
true_df = pd.read_csv(true_path)

# Add label
fake_df['label'] = 0  # Fake
true_df['label'] = 1  # Real


# Standardize subject labels in fake news to match those in true news
fake_subject_map = {
    'politics': 'politicsNews',
    'Government News': 'politicsNews',
    'left-news': 'politicsNews',
    'News': 'worldnews',
    'US_News': 'worldnews',
    'Middle-east': 'worldnews'
}

fake_df['subject'] = fake_df['subject'].map(fake_subject_map)

# Combine
df = pd.concat([fake_df, true_df], ignore_index=True)
print(df.shape)
df.head()
```

# 2 Preprocess

## 2.1 Initialize Preprocessing Tools

### 2.1.1 WordNetLemmatizer():

- A lemmatizer reduces words to their base or **dictionary** form (called lemma).

- For example, “**running**”, “**ran**”, and “**runs**” → “**run**”.

- This helps **treat variations of the same word as one feature**, **reducing dimensionality** and **improving model consistency**.

### 2.1.2 stopwords.words('english'):

- Loads a list of common English stopwords like “**the**”, “**is**”, “**and**”, “**in**”, etc.

- These words are often removed because they **carry little meaningful information** for classification tasks.

- The list is converted to a Python set for **faster lookup**.

## 2.2 Define the preprocess function

This function performs several **transformations** on the input text:

- Convert to lowercase
    - Ensures consistency by treating "**Fake**", "**FAKE**", and "**fake**" as the same word.
    
- Remove punctuation
    - Strips out characters like **. , ! ? : ;** etc.
    - Punctuation usually doesn’t help in classification and may add **noise**.
    
- Tokenize the text
    - Breaks the text into individual **words** (tokens).
    - For example: "fake news spreads fast" → ['fake', 'news', 'spreads', 'fast']
    
- Remove stopwords and lemmatize tokens
    - Filters out common stopwords.
    - Lemmatizes each remaining word to reduce it to its root form.

- Reconstruct cleaned text
    - Converts the list of cleaned tokens back into a single space-separated string.
    
## 2.3 Apply preprocessing to dataset

- Combine the article’s title and text into one column, to create a richer input representation.

- Apply the preprocess() function to every row in the df['text'] column.

The resulting cleaned, lemmatized, and token-filtered text will be used as model input.

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = str(text).lower()
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Remove punctuation
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['text'] = df['title'] + " " + df['text']
df['text'] = df['text'].apply(preprocess)
```

# 3 Add a simplified domain column



## 3.1 Define features and labels

- **X**: The input features, i.e., the preprocessed text (combined title and body).

- **y**: The target variable for classification — 0 for fake news, 1 for real news.

- domain: Stores the simplified subject label we just created (politicsNews or worldnews) — useful for stratified performance analysis.

```python
# Encode labels
X = df['text']
y = df['label']
domain = df['subject']
```

# 4 Building the Model

## 4.1 Train Test Split

This splits the dataset into:

- **80%** for training (X_train, y_train)

- **20%** for testing (X_test, y_test)

The corresponding domain labels are also split into **domain_train** and **domain_test** for later analysis.

random_state=42 ensures that the split is reproducible.

### Why split?
We train the model on one portion of the data (training set) and evaluate it on unseen data (test set) to measure generalization performance.

## 4.2 Define a machine learning pipeline

This pipeline has two components:

1. **TF-IDF Vectorizer**:
    - Converts raw text into numerical vectors based on TF-IDF (Term Frequency–Inverse Document Frequency).

    - This method **weighs** words by **how unique** or **important** they are across the dataset.

    - **max_features=5000** limits the vocabulary to the **5000 most relevant words** to **reduce dimensionality** and **overfitting**.

2. **Logistic Regression Classifier**:
    - A linear model that outputs probabilities for binary classification.

    - It’s simple, fast, and interpretable — making it **ideal for a baseline model**.

    - **max_iter=1000** ensures the optimization algorithm has enough iterations to converge.
    
### Why choose Logistic Regression?
- It’s a strong baseline for text classification.

- Works well with high-dimensional sparse data like TF-IDF vectors.

- Easy to interpret, fast to train, and requires minimal tuning.

## 4.3 Train the model
Trains the TF-IDF + Logistic Regression pipeline on the training set.

Internally, it:

- Learns the vocabulary and IDF weights from the training text.

- Fits the logistic regression model on the resulting TF-IDF vectors.

## 4.4 Make predictions and evaluate performance

**model.predict(X_test)** makes predictions on unseen test data.

**classification_report()** prints precision, recall, F1-score, and accuracy:

- **Precision**: How many predicted real articles were actually real?

- **Recall**: How many real articles were correctly identified?

- **F1-score**: Harmonic mean of precision and recall.

- **Accuracy**: Overall percentage of correct predictions.

```python
X_train, X_test, y_train, y_test, domain_train, domain_test = train_test_split(X, y, domain, test_size=0.2, random_state=42)

model = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=5000)),
    ('clf', LogisticRegression(max_iter=1000))
])

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("=== Overall Performance ===")
print(classification_report(y_test, y_pred))
```

# 5 Evaluate Model Across Different Domains

This part of the code evaluates the model’s performance across different content domains — specifically, worldnews and politicsNews. It provides a more detailed look into how well the model performs within each category.

### Purpose
While overall accuracy is useful, it doesn’t tell us where the model performs well or poorly. Since the dataset contains articles from different subject domains, we want to evaluate performance separately for each domain to:

- Check for potential bias or imbalance

- Understand which domains are harder to classify correctly

- Guide future improvements (e.g., collect more data or tune features per domain)

```python
print("=== Domain-Specific Performance ===")
domains = domain_test.unique()
for d in domains:
    idx = domain_test == d
    print(f"\n--- {d.upper()} ---")
    print(classification_report(y_test[idx], y_pred[idx]))
```

# 6 Interpret model behavior

## 6.1 get_feature_names_out()
Extracts the list of word features used by the TF-IDF vectorizer.

These are the 5000 most important words (based on TF-IDF ranking).


## 6.2 model.named_steps['clf'].coef_[0]
Gets the logistic regression coefficients for each feature.

Each number represents how strongly that word influences the prediction:

- Positive values → push prediction toward real news

- Negative values → push prediction toward fake news

## 6.3 Create a DataFrame of feature importances

Creating a DataFrame with:

- feature: the word

- importance: its corresponding logistic regression weight

Then sorts it by importance (from most positive to most negative)

## 6.4 Identify top words for each class

Top 10 fake words: highest positive weights (pushed classification toward fake)

Top 10 real words: most negative weights (pushed classification toward real)

These help us explain what the model learned.

## 6.5 Visualize word importance

Creates two horizontal bar plots:

- One showing the strongest indicators of fake news

- One showing the strongest indicators of real news

These visuals help interpret the model and answer questions like:
“What kind of language does the model associate with each class?”

```python
# Get feature names
feature_names = model.named_steps['tfidf'].get_feature_names_out()
coefficients = model.named_steps['clf'].coef_[0]

# Top words
importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': coefficients
}).sort_values(by='importance', ascending=False)

top_fake = importance_df.head(10)
top_real = importance_df.tail(10)

plt.figure(figsize=(10, 5))
sns.barplot(x='importance', y='feature', data=top_fake)
plt.title('Top Words Predicting Fake News')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x='importance', y='feature', data=top_real)
plt.title('Top Words Predicting Real News')
plt.show()
```

# 7 Confusion Matrix

```python


# --- All domains ---
cm_all = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_all, display_labels=['Fake', 'Real'])
disp.plot()
plt.title("Confusion Matrix - All Domains")
plt.show()

# --- politicsNews only ---
idx_pol = domain_test == 'politicsNews'
cm_pol = confusion_matrix(y_test[idx_pol], y_pred[idx_pol])
disp = ConfusionMatrixDisplay(confusion_matrix=cm_pol, display_labels=['Fake', 'Real'])
disp.plot()
plt.title("Confusion Matrix - politicsNews")
plt.show()

# --- worldnews only ---
idx_world = domain_test == 'worldnews'
cm_world = confusion_matrix(y_test[idx_world], y_pred[idx_world])
disp = ConfusionMatrixDisplay(confusion_matrix=cm_world, display_labels=['Fake', 'Real'])
disp.plot()
plt.title("Confusion Matrix - worldnews")
plt.show()
```

# 8 Model Improvement Using Ensemble Method
We’ll use a Random Forest classifier — an ensemble of decision trees that usually performs better than a single linear model.

```python


# New pipeline with Random Forest
improved_model = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=5000)),
    ('clf', RandomForestClassifier(n_estimators=100, random_state=42))
])

# Train and predict
improved_model.fit(X_train, y_train)
y_pred_rf = improved_model.predict(X_test)

```

# 9 Ensemble Method Evaluations

## 9.1 Evaluations Metrics

## 9.2 Confusion Matrix

```python
print("=== Improved Model - Overall Performance ===")
print(classification_report(y_test, y_pred_rf))

print("=== Improved Model - Domain-Specific Performance ===")
for d in domain_test.unique():
    idx = domain_test == d
    print(f"\n--- {d.upper()} ---")
    print(classification_report(y_test[idx], y_pred_rf[idx]))
    
# All
cm_all_rf = confusion_matrix(y_test, y_pred_rf)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_all_rf, display_labels=['Fake', 'Real'])
disp.plot()
plt.title("Confusion Matrix - All Domains (Random Forest)")
plt.show()

# Politics
cm_pol_rf = confusion_matrix(y_test[idx_pol], y_pred_rf[idx_pol])
disp = ConfusionMatrixDisplay(confusion_matrix=cm_pol_rf, display_labels=['Fake', 'Real'])
disp.plot()
plt.title("Confusion Matrix - politicsNews (Random Forest)")
plt.show()

# World
cm_world_rf = confusion_matrix(y_test[idx_world], y_pred_rf[idx_world])
disp = ConfusionMatrixDisplay(confusion_matrix=cm_world_rf, display_labels=['Fake', 'Real'])
disp.plot()
plt.title("Confusion Matrix - worldnews (Random Forest)")
plt.show()
```

# 10 Conclusion: Model Comparison and Improvement Summary
We began with a baseline logistic regression model using TF-IDF features, which achieved strong performance with 99% accuracy and F1-score across all domains. Domain-specific results showed perfect scores for worldnews, while politicsNews saw a slight drop to 98% F1-score, with a few misclassifications reflected in the confusion matrix.

To improve upon this, we implemented a Random Forest classifier as an ensemble method. This approach resulted in a **near perfect classification** performance across all metrics and domains — achieving **near 100%** accuracy, precision, recall, and F1-score, and reducing all false positives and false negatives to nearly zero. Confusion matrices confirmed the Random Forest model's ability to generalize across both politicsNews and worldnews.

This improvement demonstrates the power of ensemble methods in capturing more complex patterns in text classification tasks, making Random Forest a highly effective upgrade over the linear baseline model.

# 11 Further Improvement

For further improvement we can do augmentation, however because of time limitation, probably it will be incomplete.

```python

# Duplicate 20% of training set
augmented_df = df.sample(frac=0.2, random_state=1)

# Add a small token to simulate variation
augmented_df['text'] = augmented_df['text'] + " moreinfo"

# Combine with original dataset
df_augmented = pd.concat([df, augmented_df], ignore_index=True)
df_augmented = shuffle(df_augmented, random_state=42)

# Re-split
X_aug = df_augmented['text']
y_aug = df_augmented['label']
domain_aug = df_augmented['domain']

X_train_aug, X_test_aug, y_train_aug, y_test_aug = train_test_split(
    X_aug, y_aug, test_size=0.2, random_state=42
)

model = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=5000)),
    ('clf', LogisticRegression(max_iter=1000))
])

model.fit(X_train_aug, y_train_aug)
y_pred_aug = model.predict(X_test_aug)

print("=== Overall Performance ===")
print(classification_report(y_test_aug, y_pred_aug))
```
